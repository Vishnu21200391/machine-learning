```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Files/Item.csv")
```

```python
df
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
```

```python
df
```

```python
df['Price'].fillna(df['Price'].median(), inplace=True)
```

```python
df
```

```python
df['Bought'] = df['Bought'].fillna(df['Bought'].std())
```

```python
df['Forenoon'] = df['Forenoon'].fillna(df['Forenoon'].min())
```

```python
df['Afternoon'] = df['Afternoon'].fillna(df['Afternoon'].max())
```

```python
df
```

```python
df['Item'] = df['Item'].str.upper()
```

```python
df
```

```python
df['Bought'] = df['Bought'].astype(int)
```

```python
df
```

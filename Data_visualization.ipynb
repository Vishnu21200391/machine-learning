```python
x = [10,20,30,40,50]
y = [20,25,35,55,45]

plt.plot(x,y)
plt.title("Plotting")
plt.show()
```

```python
x = [10,20,30,40,50]
y = [20,25,35,55,45]

sns.lineplot(x=x,y=y)
plt.title("Plotting")
plt.show()
```

```python
data = np.random.rand(5, 5)  
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

corr = df.corr()
corr
```

```python
sns.heatmap(corr,annot=True, cmap="viridis", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df = pd.DataFrame({'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C', 'B', 'A']})

sns.countplot(x=df['Category'], palette="viridis")
plt.title("Category Countplot")
plt.show()
```

```python
df = pd.DataFrame({
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Values': [10, 20, 15, 25, 30, 22, 35, 40, 38]
})

sns.boxplot(x=df['Category'], y=df['Values'], palette="Set2")
plt.title("Boxplot Example")
plt.show()
```

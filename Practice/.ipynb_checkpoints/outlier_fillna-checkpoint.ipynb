```python
df['discounted_price'].fillna(df['discounted_price'].mean(), inplace=True)
```

```python
df = df.drop(['description', 'image', 'brand', 'product_specifications'], axis=1)
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

df_cleaned = remove_outliers_iqr(df, "retail_price")
df_cleaned = remove_outliers_iqr(df_cleaned, "discounted_price")

print("Original Data Size:", df.shape)
print("Cleaned Data Size:", df_cleaned.shape)


df_cleaned.to_csv("../Files/flipkart_cleaned.csv", index=False)
```

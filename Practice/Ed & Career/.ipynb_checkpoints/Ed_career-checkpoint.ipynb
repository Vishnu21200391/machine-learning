```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('education_career_success.csv')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['Age', 'High_School_GPA', 'Internships_Completed', 'Projects_Completed', 
    'Certifications', 'Soft_Skills_Score', 'Networking_Score', 'Job_Offers', 
    'Starting_Salary', 'Work_Life_Balance']] = df[['Age', 'High_School_GPA', 
    'Internships_Completed', 'Projects_Completed', 'Certifications', 'Soft_Skills_Score', 
    'Networking_Score', 'Job_Offers', 'Starting_Salary', 'Work_Life_Balance']].apply(
    lambda col: col.fillna(col.mode()[0] if not col.mode().empty else col.median()), axis=0)
```

```python
df.isnull().sum()
```

```python
df.drop(df[['Work_Life_Balance','Entrepreneurship','Certifications','Networking_Score','Career_Satisfaction','Soft_Skills_Score','High_School_GPA']], axis=1, inplace=True)
```

```python
df.info()
```

```python
Q1 = df['University_Ranking'].quantile(0.25)
Q3 = df['University_Ranking'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['University_Ranking'] >= lower_bound) & (df['University_Ranking'] <= upper_bound)]

df_cleaned.to_csv("cleaned_dataset.csv", index=False)
```

```python
data = pd.read_csv("cleaned_dataset.csv")
```

```python
data
```

```python
label_encoder = LabelEncoder()
```

```python
data['Gender'] = label_encoder.fit_transform(data['Gender'])
data['Field_of_Study'] = label_encoder.fit_transform(data['Field_of_Study'])
data['Student_ID'] = label_encoder.fit_transform(data['Student_ID'])
data['Current_Job_Level'] = label_encoder.fit_transform(data['Current_Job_Level'])
```

```python
data.head()
```

```python
corr_matrix = data.corr()
filtered_corr = corr_matrix[(corr_matrix > 0.5) | (corr_matrix < -0.5)]

plt.figure(figsize=(7, 5))
sns.heatmap(filtered_corr, annot=True, cmap="Greens", linewidths=0.5)
plt.title("Filtered Correlation Heatmap")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split

# Step 1: Create the Dataset
data = {
    "Animal": ["Dog", "Cat", "Lizard", "Elephant", "Snake"],
    "Weight (kg)": [10, 5, 1, 3000, 2],
    "Height (cm)": [150, 35, 15, 250, 101],  # Fixed typo in "Height"
    "Class": [1, 1, 0, 1, 0]
}

df = pd.DataFrame(data)

# Step 2: Define Features and Target
X = df[["Weight (kg)", "Height (cm)"]]
Y = df["Class"]

# Step 3: Train a RandomForestClassifier
rf = RandomForestClassifier(n_estimators=3, random_state=42)
rf.fit(X, Y)

# Step 4: Visualize Decision Trees
plt.figure(figsize=(15, 5))  # Adjusted figure size for better visibility

for i in range(3):
    plt.subplot(1, 3, i + 1)
    plot_tree(rf.estimators_[i], feature_names=["Weight (kg)", "Height (cm)"], filled=True, rounded=True)
    plt.title(f"Tree {i+1}")  # Fixed f-string issue

plt.tight_layout()
plt.show()

# Step 5: Visualize Feature Importance
feature_importance = rf.feature_importances_
plt.figure(figsize=(6, 4))

# Fixed feature importance visualization
sns.barplot(x=["Weight (kg)", "Height (cm)"], y=feature_importance, palette="viridis")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.title("Feature Importance in RandomForest")
plt.show()

# Step 6: Predict New Animal Type
# Convert new_animal to a DataFrame to avoid warnings
feature_names = ["Weight (kg)", "Height (cm)"]
new_animal = pd.DataFrame([[35, 100]], columns=feature_names)

# Make prediction
prediction = rf.predict(new_animal)

# Fixed Unicode escape sequences
animal_type = "Mammal \U0001F7E2" if prediction[0] == 1 else "Reptile \U0001F534"

print(f"Prediction for the new animal: {animal_type}")
```

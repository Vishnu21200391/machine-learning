```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
data = {
    'customer' : ['A', 'B', 'C', 'D', 'E', 'F'],
    'income' : [15, 16, 25, 45, 70, 85],
    'SpendingScore' : [39, 81, 6, 70, 77, 15]
}
df = pd.DataFrame(data)
print(df)
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = df[['income', 'SpendingScore']]

# Typo in parameter name: 'random_ssate' → 'random_state'
kmeans = KMeans(n_clusters=2, random_state=42)

df['cluster'] = kmeans.fit_predict(x)

plt.figure(figsize=(8, 5))

colors = ['red', 'green']

# Typo: 'color' → 'colors'
# Also, f-string formatting error: f'cluster{'cluster'}' → f'Cluster {cluster}'
for cluster in range(2):
    cluster_data = df[df['cluster'] == cluster]
    plt.scatter(cluster_data['income'], cluster_data['SpendingScore'],
                color=colors[cluster], label=f'Cluster {cluster}')

# Missing quotes around 'Centroids' label
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], color='blue', marker='x', s=200, label='Centroids')

plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('KMeans Clustering (k=2)')
plt.legend()
plt.grid(True)
plt.show()
```
